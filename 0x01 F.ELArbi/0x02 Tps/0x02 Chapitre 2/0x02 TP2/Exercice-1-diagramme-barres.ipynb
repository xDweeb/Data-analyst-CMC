# **Étape 1 : Importer les bibliothèques nécessaires**

**Nous utiliserons Seaborn pour le graphique en barres et Matplotlib pour afficher le graphique.**

```python
# Importer les bibliothèques
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Étape 2 : Charger le jeu de données tips depuis Seaborn**

```python
# Charger le jeu de données "tips"
tips = sns.load_dataset("tips")
```

# **Étape 3 : Vérifier le type du jeu de données**
**Le type du jeu de données peut être vérifié avec type().**

```python
# Afficher le type du jeu de données
print(type(tips))
```

Ce code renverra <class 'pandas.core.frame.DataFrame'>, indiquant que tips est un DataFrame, qui est le type standard pour les données tabulaires dans Pandas.

# **Étape 4 : Afficher les cinq premières lignes du jeu de données**

```python
# Afficher les cinq premières lignes
tips.head()
```

# **Étape 5 : Initialiser une figure avec une taille de 10x6 pouces**

```python
# Initialiser la figure
plt.figure(figsize=(10, 6))
```

# **Étape 6 : Tracer le graphique en barres**
**Utiliser Seaborn pour créer un graphique en barres, en spécifiant sex comme axe des X et tip comme axe des Y, et en définissant la palette de couleurs à Set2.**

```python
# Créer le graphique en barres
sns.barplot(x="sex", y="tip", data=tips, palette="Set2")
```

# **Étape 7 : Définir le titre du graphique**

```python
# Titre du graphique
plt.title("Comparaison des Pourboires entre Hommes et Femmes")
```

# **Étape 8 : Définir le titre de l’axe des X**

```python
# Titre de l'axe des X
plt.xlabel("Genre")
```

# **Étape 9 : Définir le titre de l’axe des Y**

```python
# Titre de l'axe des Y
plt.ylabel("Pourboire moyen")
```

# **Étape 10 : Afficher le graphique**

```python
# Afficher le graphique
plt.show()
```

# **Complet Code de Graphe**

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.barplot(x="sex", y="tip", data=tips, palette="Set2")
plt.title("Comparaison des Pourboires entre Hommes et Femmes")
plt.xlabel("Genre")
plt.ylabel("Pourboire moyen")
plt.show()
```
